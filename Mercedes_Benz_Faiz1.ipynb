## 4. PCA(Dimensionality Reduction)

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=0.95, random_state=11)
```

```python
train_pca=pca.fit_transform(train_feature)
test_pca=pca.transform(test_feature)
```

```python
print(train_pca.shape,test_pca.shape)
```

## Split train & Validation set from train set 

```python
from sklearn.model_selection import train_test_split
x_train,x_validate,y_train,y_validate=train_test_split(train_pca,train_target,test_size=0.25,random_state=11)
print(x_train.shape,x_validate.shape,y_train.shape,y_validate.shape)
```

## Model Building

```python
import xgboost as xgb
XGB=xgb.XGBRegressor() 
```

```python
# Finding Best Parameters for XGboost using RandomizedSearchCV
```

```python
params={'n_estimators':[50,100,200,250,300],
        'max_depth':[2,4,5,8],
        'learning_rate':[0.1,0.2,0.3],
        'min_child_weight':[1,2,3,4],
        'booster':['gbtree','gblinear'],
        'base_score':[0.25,0.5,1]   
        }
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_cv=RandomizedSearchCV(estimator=XGB, param_distributions=params, cv=4, n_jobs=4,
                             n_iter=10, verbose=5, return_train_score=True,  random_state=11)
```

```python
random_cv.fit(x_train,y_train)
```

```python
random_cv.best_estimator_
```

```python
import xgboost as xgb

XGB=xgb.XGBRegressor(base_score=0.25, booster='gbtree', 
             colsample_bylevel=1, colsample_bynode=1, colsample_bytree=1,
             early_stopping_rounds=None, enable_categorical=False,
             eval_metric=None, gamma=0, gpu_id=-1, grow_policy='depthwise',
             importance_type=None, interaction_constraints='',
             learning_rate=0.2, max_bin=256, max_cat_to_onehot=4,
             max_delta_step=0, max_depth=8, max_leaves=0, min_child_weight=4,
             missing=np.nan, monotone_constraints='()', n_estimators=50, n_jobs=0,
             num_parallel_tree=1, predictor='auto', random_state=11, reg_alpha=0,
             reg_lambda=1)
```

```python
model=XGB.fit(x_train,y_train)
y_pred=model.predict(x_validate)
print(y_pred)
```

## Evaluation Metrics

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
model.score(x_train, y_train)
```

```python
mean_absolute_error(y_validate,y_pred)
```

```python
mean_squared_error(y_validate,y_pred)
```

```python
rmse=np.sqrt(mean_squared_error(y_validate,y_pred))
rmse
```

## Forecast using Model for Test Data Set

```python
forecast=model.predict(test_pca)
forecast=pd.DataFrame(forecast,columns=['Prediction'])
forecast.head()
```
